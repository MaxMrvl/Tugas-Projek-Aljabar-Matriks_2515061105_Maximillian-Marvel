**Soal No.1**

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
```

```python
import numpy as np
```

```python
X = np.array([[12, 86, 43],
              [87, 51, 23]])
X
```

```python
#Buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
```

```python
import torch
```

```python
X_pt = torch.tensor([[56, 12, 34],
                     [56, 43, 87]])
X_pt
```

```python
#Buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
```

```python
import tensorflow as tf
```

```python
X_tf = tf.constant([[19, 91, 67],
                    [69, 42, 21]])
X_tf
```

```python
#lakukan perkalian matrix X, X_pt dan X_tf dengan suatu skalar
```

```python
skalar = 2

X_s = X * skalar
X_pt_s = X_pt * skalar
X_tf_s = X_tf * skalar

print("Matriks X (numpy) setelah perkalian dengan skalar:")
print(X_s)
print()
print("Matriks X (PyTorch) setelah perkalian dengan skalar:")
print(X_pt_s)
print()
print("Matriks X (TensorFlow) setelah perkalian dengan skalar:")
print(X_tf_s)
```
