Soal Nomor 1

```python
# buat vector x dan y dengan ukuran 1x3 menggunakan numpy
import numpy as np
```

```python
x = np.array([13, 24, 8])
x
```

```python
y = np.array([2, 11, 20])
y
```

```python
# buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
import torch
```

```python
x_pt = torch.tensor([13, 24, 8])
x_pt
```

```python
y_pt = torch.tensor([2, 11, 20])
y_pt
```

```python
# buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
import tensorflow as tf
```

```python
x_tf = tf.Variable([13, 24, 8])
x_tf
```

```python
y_tf = tf.Variable([2, 11, 20])
y_tf
```

Soal Nomor 2

```python
# hitung perkalian dot product antara vector x dan vector y
np.dot(x, y)
```

```python
# hitung perkalian dot product antara vector x_pt dan vector y_pt
torch.dot(x_pt, y_pt)
```

```python
# hitung perkalian dot product antara vector x_tf dan vector y_tf
tf.reduce_sum(tf.multiply(x_tf, y_tf))
```
