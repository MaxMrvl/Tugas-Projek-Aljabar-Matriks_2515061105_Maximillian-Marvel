```python
#Buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
```

```python
import numpy as np
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
A
```

```python
b = np.array([[1],
              [2],
              [3]])
b
```

```python
#Buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
```

```python
import torch as pt
```

```python
A_pt = pt.tensor([[5, 4, 6],
                  [8, 6, 5]])
A_pt
```

```python
b_pt = pt.tensor([[4],
                  [2],
                  [7]])
b_pt
```

```python
#Buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
```

```python
import tensorflow as tf
```

```python
A_tf = tf.Variable([[5, 7, 8],
                    [4, 5, 3]])
A_tf
```

```python
b_tf = tf.Variable([[1],
                    [2],
                    [9]])
b_tf
```
